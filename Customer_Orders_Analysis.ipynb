## Top Customers
- Find the top 10 customers

```python
top_customers = Sales.groupby('Customer_Name')['Sales'].sum().nlargest(10)

plt.figure(figsize=(12,6))
top_customers.plot(kind='bar', color='skyblue')
plt.title("Top 10 Customers by Sales")
plt.xlabel("Customer_Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

## Monthly Sales Trend
- Check how sales grow/shrink across months.

```python
Sales['Order_Date'] = pd.to_datetime(Sales['Order_Date'])

Sales['Month'] = Sales['Order_Date'].dt.to_period('M')

monthly_sales = Sales.groupby('Month')['Sales'].sum()

plt.figure(figsize=(12,6))
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

## Category wise Sales
- Find the Sales percentage by Category

```python
category_sales = Sales.groupby('Category')['Sales'].sum()

plt.figure(figsize=(6,6))
category_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99'])
plt.title("Sales Distribution by Category")
plt.ylabel("")
plt.show()
```

## Ship_Mode
- Find Ship_Mode by Region

```python
sns.countplot(x = 'Region', hue = 'Ship_Mode', data = Sales)
```

## Segment
- Identify Region wise Segment Distribution

```python
sns.countplot(x = 'Region', hue = 'Segment', data = Sales)
```

## Region
- Identify Region Wise Sales

```python
Sales_count = Sales['Region'].value_counts().head(10)
plt.figure(figsize=(7,4))
Sales_count.plot(kind="bar", color="#FF6F61")
plt.title("Top Sales Distribution")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

## Sub_Category
- Find Top Sub_Category Distribution

```python
Category = Sales['Sub_Category'].dropna().str.split(',').explode().str.strip().value_counts()
plt.figure(figsize=(7,4))
Category.plot(kind="barh", color="#009B77")
plt.title("Top Sub_Category")
plt.tight_layout()
plt.show()
```
